Алгоритм MinMaxScaler является методом масштабирования, который линейно преобразует данные так, чтобы они помещались в заданный диапазон, обычно между 0 и 1. Это достигается путем вычитания минимального значения признака из каждого значения, а затем деления разности на размах значений (максимальное значение минус минимальное значение). Преобразование можно выразить следующей формулой:

$X_{\text{scaled}} = \frac{X - X_{\min}}{X_{\max} - X_{\min}}$

Где:
- $X$ — исходное значение признака,
- $X_{\min}$ — минимальное значение этого признака в наборе данных,
- $X_{\max}$ — максимальное значение признака в наборе данных,
- $X_{\text{scaled}}$ — масштабированное значение признака.

Если требуется масштабировать данные в другой диапазон, например, $[a, b]$, то формула модифицируется:

$X_{\text{scaled}} = a + \frac{(X - X_{\min}) \times (b - a)}{X_{\max} - X_{\min}}$

````python
new_dataset = np.array([])
print(dataset.shape)
for i in range(dataset.shape[1]):
  max_element = np.max(dataset[:, i])
  min_element = np.min(dataset[:, i])
  new_dataset = np.append(new_dataset, (dataset[:, i] - min_element)/(max_element - min_element))
new_dataset = np.reshape(new_dataset, (dataset.shape))
print(new_dataset)

#проверка:
print(np.max(new_dataset))
print(np.min(new_dataset))
````

### **Задание №1. Реализация модели линейной регрессии, используя объектно-ориентированный подход:**

Создайте класс `LinearRegression`, который реализует модель линейной регрессии для одной или нескольких независимых переменных.

#### **Теоретическая часть:**

1. **Линейная регрессия** - это метод для моделирования отношения между скалярной зависимой переменной `y` и одной или несколькими независимыми переменными (или "предикторами") `X`.

2. **Гипотеза линейной регрессии** представляется в виде:
   
   $ \hat{y} = h_{\theta}(X) = \theta^TX $

   где:
   - $\hat{y}$ - предсказанное значение,
   - $X$ - вектор признаков (с добавленным нулевым элементом, равным 1, для учета свободного члена),
   - $\theta$ - вектор параметров модели, включая свободный член и коэффициенты при признаках.

3. **Метод наименьших квадратов** используется для нахождения оптимальных параметров $\theta$. Цель - минимизировать сумму квадратов ошибок между предсказанными и реальными значениями.

   Функция стоимости $J(\theta)$:
   
   $ J(\theta) = \frac{1}{2m} \sum_{i=1}^{m} (h_{\theta}(X^{(i)}) - y^{(i)})^2 $

   где:
   - $m$ - количество обучающих примеров,
   - $X^{(i)}$ - $i$-й обучающий пример,
   - $y^{(i)}$ - реальное значение для $i$-го обучающего примера.

4. **Аналитическое решение** для нахождения $\theta$:

   $ \theta = (X^TX)^{-1}X^Ty $

Процесс вычисления коэффициентов простой линейной регрессии с помощью аналитического решения можно продемонстрировать на простом примере. Предположим, у нас есть набор данных, где $X$ - это независимая переменная, а $y$ - зависимая переменная.

Давайте возьмем простой набор данных:
$ X = \begin{bmatrix} 1 & 1 \\ 1 & 2 \\ 1 & 3 \end{bmatrix}, y = \begin{bmatrix} 1 \\ 2 \\ 2 \end{bmatrix} $

Здесь первый столбец $X$ - это всегда 1 (для учета свободного члена), а второй столбец - это значения независимой переменной.

1. Вычисляем $X^T$ (транспонируем матрицу $X$):
$ X^T = \begin{bmatrix} 1 & 1 & 1 \\ 1 & 2 & 3 \end{bmatrix} $

2. Вычисляем произведение $X^TX$:
$ X^TX = \begin{bmatrix} 3 & 6 \\ 6 & 14 \end{bmatrix} $

3. Находим обратную матрицу к $X^TX$:
$ (X^TX)^{-1} = \frac{1}{(3*14 - 6*6)} \begin{bmatrix} 14 & -6 \\ -6 & 3 \end{bmatrix} = \frac{1}{42-36} \begin{bmatrix} 14 & -6 \\ -6 & 3 \end{bmatrix} = \begin{bmatrix} 14/6 & -1 \\ -1 & 1/2 \end{bmatrix} $

4. Вычисляем произведение $X^Ty$:
$ X^Ty = \begin{bmatrix} 5 \\ 11 \end{bmatrix} $

5. Наконец, вычисляем $\theta$ по формуле:
$ \theta = (X^TX)^{-1}X^Ty = \begin{bmatrix} 14/6 & -1 \\ -1 & 1/2 \end{bmatrix} \begin{bmatrix} 5 \\ 11 \end{bmatrix} = \begin{bmatrix} (14*5 - 11*6)/6 \\ (-5 + 11/2) \end{bmatrix} = \begin{bmatrix} 4/6 \\ 1/2 \end{bmatrix} = \begin{bmatrix} 2/3 \\ 1/2 \end{bmatrix} $

Таким образом, уравнение нашей линейной регрессии будет выглядеть как $y = 2/3 + 1/2x$.

Обратите внимание, что при реальном использовании матричных операций и обратных матриц следует быть внимательным к условиям их применимости, а также к точности вычислений в случае работы с большими наборами данных или плохо обусловленными матрицами.

#### **Простыми словами:**

4. **Метод `score`:**
   - Реализуйте метод `score`, который возвращает коэффициент детерминации $R^2$, показывающий, насколько хорошо предсказания соответствуют реальным значениям.

    Коэффициент детерминации, также известный как $R^2$, измеряет долю дисперсии зависимой переменной, которая может быть предсказана на основе независимой переменной(ых) в модели линейной регрессии. Формула для $R^2$ выглядит следующим образом:

    $ R^2 = 1 - \frac{SS_{res}}{SS_{tot}} $,

    где:
    - $SS_{res}$ — сумма квадратов остатков (остаточная сумма квадратов), $ SS_{res} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $
    - $SS_{tot}$ — общая сумма квадратов (сумма квадратов отклонений от среднего значения зависимой переменной), $ SS_{tot} = \sum_{i=1}^{n} (y_i - \bar{y})^2 $
    - $y_i$ — фактические значения,
    - $\hat{y}_i$ — предсказанные значения,
    - $\bar{y}$ — среднее значение $y$.
    ```python
    # Пример входных данных
    y_true = np.array([1, 2, 3, 4, 5])
    y_pred = np.array([1.1, 1.9, 3.0, 4.1, 4.9])
    # Пример выходных данных
    R^2 Score: 0.996
    ```
    Коэффициент детерминации изменяется в диапазоне от −∞ до 1. **Если он равен 1, это соответствует идеальной модели**, когда все точки наблюдений лежат точно на линии регрессии, т.е. сумма квадратов их отклонений равна 0. **Если коэффициент детерминации равен 0, это означает, что связь между переменными регрессионной модели отсутствует**, и вместо нее для оценки значения выходной переменной можно использовать простое среднее ее наблюдаемых значений.



#### **Ваша реализация:**

````python
import numpy as np

y_true = np.array([1, 2, 3, 4, 5])
y_pred = np.array([1.1, 1.9, 3.0, 4.1, 4.9])

def found_R_in_2(a, b):
  SS_res = np.sum((a - b) ** 2)
  y_sred = np.mean(a)
  SS_tot = np.sum((a - y_sred) ** 2)
  R_in_2 = 1 - (SS_res/SS_tot)
  print('R^2 Score: ', R_in_2)
found_R_in_2(y_true, y_pred)


class LinearRegression:
    def init(self):
      self.theta = None

    def fit(self, X, y):
      self.X = X
      self.y = y
      self.X = np.c_[np.ones(self.X.shape[0]), self.X]
      X = self.X
      self.theta = np.array(np.linalg.pinv(X.T @ X) @ (X.T @ y))
      return self.theta

    def predict(self, X):
      X = np.c_[np.ones(X.shape[0]), X]
      return X @ self.theta

    def score(self, X, y):
      found_R_in_2(X, y)
````


### **Задание 2: Реализация алгоритма PCA с использованием NumPy**

Метод главных компонент (Principal Component Analysis, PCA) — это статистическая процедура, использующая ортогональное преобразование для конвертации набора наблюдений возможно коррелирующих переменных в набор значений линейно некоррелирующих переменных, называемых главными компонентами. Основная идея состоит в том, чтобы уменьшить размерность данных, сохраняя при этом как можно больше информации.

Главные компоненты выстраиваются так, что первая главная компонента имеет наибольшую возможную дисперсию (то есть, учитывает как можно больше информации о структуре данных), и каждая следующая компонента в свою очередь имеет наибольшую дисперсию при условии, что она ортогональна к предыдущим компонентам.

#### **Простыми словами:**

> Метод главных компонент (PCA) — это как будто вы решаете упростить сложную книгу до нескольких предложений, сохраняя при этом её основной смысл. В мире данных, где информация может быть сложной и перегруженной, PCA помогает выделить самое важное, убирая лишние детали.
>
> Представьте, что у вас есть огромная таблица с множеством данных о людях: рост, вес, возраст, доход, расходы на еду, расходы на одежду и так далее. Всё это может быть перегружающим если вы хотите понять общие закономерности. Как упростить всё это до нескольких ключевых показателей, не теряя при этом важную информацию?
>
> Здесь на помощь приходит PCA. Этот метод "сжимает" множество переменных (например, разные виды расходов) в меньшее количество новых переменных, называемых главными компонентами. Эти компоненты выбираются таким образом, чтобы как можно лучше отражать структуру и взаимосвязи в исходных данных. В итоге, используя PCA, вы можете уменьшить сложность данных и сделать их проще для анализа и визуализации, при этом сохраняя важные закономерности и тренды.
>
> Важно отметить, что PCA работает на предположении, что важные аспекты данных часто проявляются через их вариативность; то есть, чем больше данные варьируются в определённом направлении, тем важнее это направление считается.
>
> Таким образом, PCA — это как фильтр, который помогает сосредоточить внимание на самом важном в большом объёме данных, убирая "шум" и упрощая анализ.

2. Функция должна центрировать данные, вычислить ковариационную матрицу, собственные значения и собственные векторы.
3. Отсортировать собственные векторы по убыванию собственных значений и выбрать первые `n_components` векторов для формирования матрицы проекции.
4. Преобразовать исходные данные, умножив их на матрицу проекции.
5. Вернуть преобразованные данные и матрицу проекции.

### **Реализация метода главных компонент (PCA):**






#### 1. Центрирование данных:

Первым делом необходимо центрировать данные, то есть вычесть из каждого признака его среднее значение. Это делается для того, чтобы среднее значение каждого признака в новых данных равнялось нулю.

Формула для центрирования данных:
$ X_{centered} = X - \text{mean}(X) $,

где $ \text{mean}(X) $ - это вектор средних значений каждого столбца матрицы $ X $.

#### 2. Вычисление ковариационной матрицы

Ковариационная матрица показывает, как изменяются вместе пары признаков. Для центрированных данных ковариационную матрицу можно вычислить по формуле:

$ \Sigma = \frac{1}{n-1} X_{centered}^T X_{centered} $

где $ \Sigma $ - ковариационная матрица, $ n $ - количество наблюдений.



#### 3. Вычисление собственных значений и собственных векторов ковариационной матрицы

Собственные значения и собственные векторы ковариационной матрицы показывают направления наибольшей дисперсии данных. Собственные векторы используются для проекции данных в новое пространство.

$ \Sigma v = \lambda v $

где $ \lambda $ - собственное значение, $ v $ - собственный вектор.



#### 4. Сортировка собственных векторов

Собственные векторы сортируются по убыванию соответствующих им собственных значений. Первые `n_components` собственных векторов будут составлять матрицу проекции.



#### 5: Преобразование данных

Исходные данные проецируются на новое пространство, умножаясь на матрицу проекции.

$ X_{proj} = X_{centered} P $

где $ P $ - матрица проекции, составленная из выбранных собственных векторов.

````python
import numpy as np

def PCA(X, n_components):
    # Шаг 1: Центрирование данных
    X_mean = np.mean(X, axis = 0)
    X_c = X - X_mean
    # Шаг 2: Вычисление ковариационной матрицы (np.cov(arg, rowvar=False))
    Cov_matrix = np.cov(X_c, rowvar = False)
    # Шаг 3: Вычисление собственных значений и векторов (напр. с помощью np.linalg.eig())
    eigen_values, eigen_vectors = np.linalg.eig(Cov_matrix)
    # В случае возникновения ошибок читайте документацию по данному методу
    # Шаг 4: Сортировка собственных векторов
    index_sorted_eigen_values = np.argsort(eigen_values)[::-1]
    sorted_eigen_values = np.take(eigen_values, index_sorted_eigen_values)
    a = []
    eigen_vectors = eigen_vectors.T
    for i in range(eigen_vectors.shape[0]):
      a.append(eigen_vectors[index_sorted_eigen_values[i]])
    index_sorted_eigen_vectors = np.array(a)
    index_sorted_eigen_vectors.T
    projection_matrix = index_sorted_eigen_vectors[:, :n_components]
    # Шаг 5: Преобразование данных
    X_projected = X_c @ projection_matrix
    return X_projected, projection_matrix
````
